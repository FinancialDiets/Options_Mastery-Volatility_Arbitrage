```python
from volatility_arbitrage.pricing_model.heston_model import (
    Correlation,
    HestonParams,
    generate_inefficient_market,
)
import matplotlib.pyplot as plt
import numpy as np
```

```python
corr = Correlation(-0.754, -0.4, 0.6)
imp_var_params = HestonParams(kappa=5.13, mean_of_var=0.044, vol_of_var=0.52)
real_var_params = HestonParams(kappa=5.13, mean_of_var=0.16**2, vol_of_var=0.3)

lr, real_var, imp_var = generate_inefficient_market(
    real_var_params=real_var_params,
    imp_var_params=imp_var_params,
    corr=corr,
    num_path=100,
    length=252,
    time_delta=1 / 252,
)
```

```python
plt.plot(np.cumsum(lr, axis=0))
plt.show()
```

```python
plt.plot(np.sqrt(real_var))
plt.show()
```

```python
plt.plot(np.sqrt(imp_var))
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(np.sqrt(real_var[:, 0]))
ax.plot(np.sqrt(imp_var[:, 0]))

ax1 = ax.twinx()
ax1.plot()
```
